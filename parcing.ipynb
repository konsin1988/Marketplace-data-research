```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver import ChromeOptions
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from time import sleep
import pandas as pd

options = ChromeOptions()
# options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--remote-debugging-port=9222')

women_hoodie = []

driver = webdriver.Chrome(options=options)
driver.get('https://www.lamoda.ru/women-home/')
sleep(2)
try:
    input_box = driver.find_element(By.CSS_SELECTOR, 'input')
    input_box.send_keys('Женское худи')
    input_box.send_keys(Keys.RETURN)
    
    sleep(5)
    

    for i in range(len(driver.find_elements(By.CLASS_NAME, 'x-product-card__pic-catalog'))):
        href = driver.find_elements(By.CLASS_NAME, 'x-product-card__pic-catalog')
        href[i].click()
        sleep(3)
    
        title = driver.find_element(By.CLASS_NAME, 'ui-product-brand-title').text
        product_type = driver.find_element(By.CLASS_NAME, '_modelName_mnqvr_21').text
        product_art = driver.find_element(By.CLASS_NAME, 'ui-product-description-attribute-sku').text
        price = int(driver.find_element(By.CLASS_NAME, '_price_aaf9t_11').text.strip('₽').replace(' ', ''))
        women_hoodie.append([title, product_type, product_art, price])
    
        driver.back()
        sleep(3)
    
    
except Exception as e:
    print(f'Error: {e}')
    driver.quit()
else:
    driver.quit()

pd.DataFrame(women_hoodie, columns = ['brand', 'product_type', 'code', 'price'])
```

## Get data group by category, gender (over 1500 samples)

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver import ChromeOptions
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from time import sleep

women_hoodie = []

options = ChromeOptions()
# options.add_argument('--headless')
options.add_argument('--disable-infobars')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--no-sandbox')
options.add_argument('--remote-debugging-port=9222')

driver = webdriver.Chrome(options=options)
driver.get('https://www.lamoda.ru/women-home')
sleep(3)


try:
    for category in ['Худи', 'Свитшот', 'Жилетка', 'Рубашка']:
        for gender in ['female', 'male']:
            search_option = driver.find_element(By.CSS_SELECTOR, 'input')
            search_option.send_keys(Keys.CONTROL + "a")
            search_option.send_keys(Keys.DELETE)
            search_option.send_keys(category)
            search_option.send_keys(Keys.RETURN)
            sleep(5)
            gender_target_class = '_women_2tc05_34' if gender == 'female' else '_men_2tc05_26'
            for_gender = driver.find_element(By.CLASS_NAME, gender_target_class)
            for_gender.click()
            sleep(4)
            for _ in range(2):
                for i in range(len(driver.find_elements(By.CLASS_NAME, 'x-product-card__pic-catalog'))):
                    href = driver.find_elements(By.CLASS_NAME, 'x-product-card__pic-catalog')
                    product_link = href[i].get_attribute('href')
                    href[i].click()
                    sleep(4)
        
                    brand = driver.find_element(By.CLASS_NAME, 'ui-product-brand-title').text
                    product_name = driver.find_element(By.CLASS_NAME, '_modelName_mnqvr_21').text
                    product_art = driver.find_element(By.CLASS_NAME, 'ui-product-description-attribute-sku').text
                    price = driver.find_element(By.CLASS_NAME, '_price_aaf9t_11').text.strip('₽').replace(' ', '')
                    women_hoodie.append([product_art, gender, category, brand, product_name, price, product_link])
                
                    driver.back()
                    sleep(5)
                
                next_button = driver.find_element(By.CLASS_NAME, '_button_6v5hm_16')
                next_button.click()
                sleep(5)

except Exception as e:
    print(f'Error: {e}')
    driver.quit()
else:
    driver.quit()
```
